```r
#vectopattern: [vector-of int] string -> string
#concentates [vector-of x] to a single string using `delim' as delimiter
vectopattern<-function(ids, delim = "|"){
    
  for (i in 1:length(ids)){
     if (i == 1){
         string_ = ids[i]
     }else if(i<length(ids)){
        string_ = paste(string_," ",ids[i], sep = "") 
     }else{
        string_ = paste(string_," ",ids[i],"\n", sep = "") 
     }
  }
  return (string_)
}
```

```r
# Converts a dataframe to the timeseries format specified by `bnstruct`                                                   
get.timesSeries<- function(data, df.names, FUN){                
    # Takes `replicates` and just outputs generic names (rep1, rep2, ... , repk) for each of the replicates                                             
    get.ROWnames <-(function(replicates) (purrr::map_chr(1:length(replicates), (function(i) (paste("rep",i,sep=""))))))

    # Takes all the names of the genes and appends _t to them to indicate the time interval                                                      
    get.COLnames<- (function(genes, t) (purrr::map_chr(genes, (function(nm) (paste(nm,"_t",t,sep=""))))))
    
    # orginal gene names
    genes = rownames(data)
    discetenes = rep(FALSE,length(genes))  
    node.sizes = rep(3,length(genes)) 
                                                        
    #loop over names concenate into conforming dataframe w/ generic lables
    for (t in 1:length(df.names)){
        print(t)
        oldNames <- FUN(unlist(df.names[t])) #Replace w/ FUN and df.names
        df = data[,oldNames]            
        rNames <- get.ROWnames(oldNames)
        cNames <- get.COLnames(rownames(df),t)
            
        # Take the transpose of the dataframe and standardize the names. 
        df = t(df)
        colnames(df) = cNames
        rownames(df) = rNames
        if (t == 1){
            df.output = df
        }else{
            df.output = rowr::cbind.fill(df.output,df, fill=NA)
        }
    }
    
    # write the data   
    out = file(file.path(cleandata.dir,"gene.data"), open = 'w')
    df.output[is.na(df.output)] <- "?"
    writeLines(apply(df.output,1,vectopattern), out, sep="")
    close(out)
                                                        
    # write the header
    out = file(file.path(cleandata.dir,"gene.header"), open ='w')
    cat(vectopattern(genes),file=out)
    print(length(genes))
    close(out)
    
    out = file(file.path(cleandata.dir,"gene.header"),open = 'a')
    cat(vectopattern(node.sizes),file=out)
    close(out)
                                                        
    out = file(file.path(cleandata.dir,"gene.header"), open ='a')
    cat(vectopattern(discetenes),file=out)
    close(out)
}

# Converts continous df to a discrete df, discretize w/in rows
continous.to.discrete<-function(df, k = 3){
    colNAME = colnames(df)
    rowNAME =  rownames(df)
    return.df = arules::discretizeDF(data.frame(t(df)), default = list(method = "cluster", breaks = k))
    return.df = sapply(return.df, as.numeric)
    return.df = t(return.df)
    colnames(return.df) = colNAME
    rownames(return.df) = rowNAME
    return (return.df)
}  
                                                        
########## Parameters needed to format data into proper timeseries format ###########3                                                        

# Sometimes the names don't match the actual names for instance instead of "3Months_9" we have "X3Months_9"
# so this just appends "X" to the name
convert.namesAH <- (function(df) (purrr::map_chr(df, (function(obs) (paste("X",obs,sep=""))))))

#ordered names of the columns of the microarray daraset
ah.names <- list(ah.names.3Months,ah.names.6Months,ah.names.9Months, ah.names.12Months,ah.names.23Months)                                  

                                               
# convert from continous to discrete of 10 bins using k-means to determing groupings                                             
# test = 
                                                      
# convert to proper time series format.                                                
get.timesSeries(continous.to.discrete(alzheimer.dta[1:100,], k = 3),ah.names, convert.namesAH)                                  
              
# names of objects to construc to B-object                                              
dta.fname = file.path(cleandata.dir,"gene.data")
header.fname =  file.path(cleandata.dir,"gene.header")                                               
layering = layering=c(rep(1, 100),
                      rep(2, 100),
                      rep(3, 100),
                      rep(4, 100),
                      rep(5, 100))                       
                                               
# get the BNDdataset                                               
dataset<- BNDataset(dta.fname, header.fname, num.time.steps =5,starts.from=1)    
dataset <- impute(dataset)                                                      
#bds.bootstrap = bootstrap(bds, num.boots = 50, imputation = FALSE)                                                                                           
dbm <- learn.dynamic.network(dataset,num.time.steps = 5, algo = "sem", scoring.func = "BIC")
```
